```python
%pylab inline
from scipy.fftpack import fft, ifft
import scipy.signal as sig
from scipy.io.wavfile import read, write
```

```python
[sam_freq,data]=read('g1.wav')
T=1/sam_freq
print(sam_freq)
duration = len(data)/sam_freq

n=linspace(0,duration, num=int(sam_freq*duration))
print(len(n))

M=round(sam_freq)
plt.figure(figsize = [15,5])
plt.subplot(1,2,1)
#Complete time graph
xlabel('sample time')
ylabel('amplitude')
title('Time signal')

plot(n,data)

plt.subplot(1,2,2)

#Time domain plot in range  0 to 0.03 sec
xlabel('sample time')
ylabel('amplitude')
title('Zoomed time domain plot')

xlim(0 ,0.03)
plot(n,data)
```

```python
#frequency domain graph
freq_data= fft(data)
freq_data2 = abs(freq_data[0:M])

freq_xAxis1=linspace(0.0,1/(2*T),M)

plt.figure(figsize =[15,5])
plt.subplot (1,2,1)

xlabel('Frequency')
ylabel('Amplitude')
title('Frequency Domain graph')
plot(freq_xAxis1,freq_data2)
print(len(freq_data2))
grid()

#Frequency domain graph on range 1000 to 2000
plt.subplot(1,2,2)
xlabel('Frequency')
ylabel('amplitude')
title('Frequency domain graph on range 0 to 2000')
xlim(100,1000)
plot(freq_xAxis1,freq_data2)
grid()
#from the plot we observe 3 major frequencies at 600,650,900 freq_xAxis
#other values are regared as noise
```

```python
# we get cutoff frequencies 
fc1=525
fc2=600
fc3=675
fbp=fc3-20
fbs=fc1+20
# noramlizing
freq_norm1 = 2*fc1/sam_freq
freq_norm2 = 2*fc2/sam_freq
freq_norm3 = 2*fc3/sam_freq
wcbp=2*fbp/sam_freq
wcbs=2*fbs/sam_freq
```

# LOW PASS (LP) FILTER DESIGN

```python
#LOWPASS filter at fc=1200
m = 1000 #Window length/order of the filter. 

blp = sig.firwin(m,freq_norm1,window='black') #low pass
[w,Hklp] = sig.freqz(blp, worN = sam_freq)

freq_xAxislp = sam_freq * w/(2*pi)
hklp2 = abs(Hklp)
xlim(0,5000)
xlabel('Frequency')
ylabel('amplitude')
title('Low pass filter design (H(jw))')
plot(freq_xAxislp,hklp2)
grid()
```

```python
#LOWPASS filter 
#convolution in time domain is multiplication in frequency domain
#filter * audio signal both in frequency domain
print(len(hklp2))
print(len(freq_data2))
print(sam_freq)
outlp = hklp2[0:44100] * freq_data2
xlim(0,5000)
plot(freq_xAxislp[0:44100],abs(outlp))
xlabel('Frequency')
ylabel('amplitude')
title('Applied LowPass filter')
grid()
```

```python
#Frequency domain graph on range 1000 to 2000

xlabel('Frequency')
ylabel('amplitude')
title('Frequency domain graph on range 0 to 2000')
xlim(0,2000)
plot(freq_xAxislp[0:44100],abs(outlp))
grid()
```

```python
plt.figure(figsize=[15,5])
plt.subplot(1,2,1)
outlp_time=ifft(outlp)
xlabel('sample time')
ylabel('amplitude')
title('Time domain of LowPass-filter-applied signal')
plot(n[0:44100],outlp_time)
grid()

#plot low pass output in time doamin range 0 to o.o1

plt.subplot(1,2,2)
xlim(0,0.03)
xlabel('sam_freq time')
ylabel('amplitude')
title('Low pass output in time doamin in range 0 to o.o3 (zoomed)')
plot(n[0:44100], outlp_time)
grid()
```

```python
# save output as audio file
write('g1_FIR_LP.wav',round(sam_freq/2),abs(outlp_time))
```

# HIGH PASS FIR

```python
# setting up highpass 
bhp = sig.firwin(m+1,freq_norm3, pass_zero = False,window='rect') #highpass
[w,Hkhp] = sig.freqz(bhp, worN = sam_freq)
freq_xAxishp = sam_freq * w/(2*pi)
hkhp2 = abs(Hkhp)

xlabel('Frequency')
ylabel('amplitude')
title ('HighPass filter design(H(jw))')
xlim(0,5000)
plot(freq_xAxishp,hkhp2)
grid()
```

```python
#applying the HIGHPASS filter 
#convolution in time domain is multiplication in frequency domain
#filter * audio signal both in frequency domain

outhp = hkhp2[0:44100] * freq_data2
xlim(0,5000)
plot(freq_xAxishp[0:44100],abs(outhp))
xlabel('Frequency')
ylabel('amplitude')
title('Frequency domain of HighPass-filter-applied signal')
grid()
```

```python

xlabel('Frequency')
ylabel('amplitude')
title('Frequency domain graph on range 0 to 2000')
xlim(0,2000)
plot(freq_xAxishp[0:44100],abs(outhp))
grid()
```

```python
plt.figure(figsize = [15,5])
plt.subplot(1,2,1)
outhp_time=ifft(outhp)
xlabel('sam_freq time')
ylabel('amplitude')
title('Time domain of HighPass-filter-applied signal')
plot(n[0:44100],outhp_time)
grid()

#plot high pass output in time doamin in range 0 t0 0.01
plt.subplot(1,2,2)
xlim(0,0.01)
xlabel('sam_freq time')
ylabel('amplitude')
title('Time domain of HighPass-filter-applied signal in range 0 to 0.01 (zoom)')
plot(n[0:44100],outhp_time)
grid()
```

```python
# save output as audio file
write('g1_FIR_HP.wav',round(sam_freq/2),abs(outhp_time))
```

# BAND PASS

```python
bbp = sig.firwin(M,[freq_norm1,freq_norm3], pass_zero =False,window='hann')#bandpass
[wbp,Hkbp] = sig.freqz(bbp, worN = sam_freq)
freq_xAxisbp = sam_freq * wbp/(2*pi)
hkbp2 = abs(Hkbp)
hk_db = 20* log(hkbp2)


xlim(0,1200)
xlabel('Frequency')
ylabel('amplitude')
title('BandPass filter design (H(jw))')
plot(freq_xAxisbp,hkbp2)

grid()
```

```python
#applying the BANDPASS filter 
#convolution in time domain is multiplication in frequency domain
#filter * audio signal both in frequency domain
xlim(0,0.01)
xlabel('sam_freq time')
ylabel('amplitude')


outbp = hkbp2[0:44100] * freq_data2
xlim(0,4000)
plot(freq_xAxisbp[0:44100],abs(outbp))
xlabel('Frequency')
ylabel('amplitude')
title('Frequency Domian BandPass-filter-appied signal')
grid()
```

```python
#reverse output and plot in time domain 
plt.figure(figsize= [15,5])
plt.subplot(1,2,1)

outbp_time=ifft(outbp)
xlabel('sam_freq time')
ylabel('amplitude')
title('Time Domian BandPass-filter-appied signal')
plot(n[0:44100],outbp_time)
grid()

#Zoomed plot band pass output in time doamin
plt.subplot(1,2,2)
xlim(0,0.01)
xlabel('sam_freq time')
ylabel('amplitude')
title('Zoomed Time Domian BandPass-appied signal')
plot(n[0:44100],outbp_time)
grid()
```

```python
# save output as audio file
write('g1_FIR_BP.wav',round(sam_freq/2),abs(outbp_time))
```

# IIR FILTER

# LOW PASS IIR

```python
#LOWPASS FILTER DESIGN IIR
n = 5 # ORDER OF DESIGN

[blp_iir,a] = sig.butter(n,freq_norm1, btype='lowpass') #lowpass 
[wlp_iir,hklp_iir] = sig.freqz (blp_iir,a, worN = sam_freq)
freq_xAxislp_iir = sam_freq * wlp_iir/(2*pi)
hklp_iir2 = abs(hklp_iir)

xlim(0,5000)
xlabel('Frequency')
ylabel('amplitude')
title('low pass filter design')
plot(freq_xAxislp_iir,hklp_iir2)
grid()
```

```python
#applying the LOWPASS filter 
#convolution in time domain is multiplication in frequency domain
#filter * audio signal both in frequency domain
outlp_iir = hklp_iir2[0:44100] * freq_data2
xlim(0,5000)
plot(freq_xAxislp_iir[0:44100],abs(outlp_iir))
xlabel('Frequency')
ylabel('amplitude')
title('frequency response graph')
grid()
```

```python
#reverse output and plot in time domain 
plt.figure(figsize = [15,5])
plt.subplot(1,2,1)
outlp_iir_time=ifft(outlp_iir)
xlabel('sam_freq time')
ylabel('amplitude')
title('reversw out put in time domain')
n=linspace(0,round(duration),round(sam_freq*duration))
#print(n)
plot(n[0:44100],outlp_iir_time)
grid()

#Zoomed plot  output in time doamin
plt.subplot(1,2,2)
xlim(0,0.01)
title('out put in range 0 to 0.01(zoom)')
xlabel('sam_freq time')
ylabel('amplitude')
plot(n[0:44100],outlp_iir_time)
grid()
```

```python
# save output as audio file
write('g1_IIR_LP.wav',round(sam_freq/2),abs(outlp_iir_time))
```
